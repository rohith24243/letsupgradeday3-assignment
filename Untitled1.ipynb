# 2. program for print prime numbers in between 1 to 1000 using for loop

```python
lower =1
upper =1000

print("Prime numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```

# 3. program of printing the tables from 1,10 using nested for loops

```python
for i in range(1,11):
    print("\n\nMULTIPLICATION TABLE FOR %d\n" %(i))
    for j in range(1,11):
        print(i, 'x', j, '=', i*j)
```
